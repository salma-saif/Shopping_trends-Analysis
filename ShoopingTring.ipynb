```python

cat_columns = [
 'Gender','Category','Size','Season','Subscription Status','Promo Code Used'
]



# Create a figure for the pie charts
plt.figure(figsize=(12, 10))

# Loop through each one-hot encoded EducationField column and its corresponding name
for i, col in enumerate(cat_columns):

    level_data = data[col].value_counts()

    # Plot the pie chart
    plt.subplot(2, 3, i+1)
    plt.pie(level_data, labels=level_data.index, autopct='%1.1f%%', startangle=140)

    # Title for each subplot
    plt.title(f'Distribution of {col}')

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()
```

```python
# Step 1: Create a color map (dictionary: color name → actual color code)
unique_colors = data['Color'].unique()

# Optional: Try matching color names to real color codes
import matplotlib.colors as mcolors
named_colors = mcolors.CSS4_COLORS  # Built-in CSS color names

# Match color names from data to CSS4 colors
color_palette = {color: named_colors.get(color.lower(), 'gray') for color in unique_colors}

# Step 2: Plot with manual color palette
plt.figure(figsize=(14, 6))
sns.countplot(x='Season', hue='Color', data=data, palette=color_palette)
plt.title('Distribution of Colors per Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend(title='Color', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Define frequency mapping per month
frequency_map = {
    'Daily': 30,
    'Bi-Weekly': 2,   # every 2 weeks = 2 per month
    'Weekly': 4,
    'Fortnightly': 2,  # same as Bi-Weekly
    'Monthly': 1,
    'Quarterly': 1/3,
    'Annually': 1/12
}
```

```python
# Map textual values to numeric
data['Frequency_numeric'] = data['Frequency of Purchases'].map(frequency_map)
```

```python
avg_frequency = data['Frequency_numeric'].mean()
print("Average frequency of purchases per customer (per month):", round(avg_frequency, 2))
```

```python
avg_per_year = avg_frequency * 12
print("Average frequency per customer per year:", round(avg_per_year, 1))
```
